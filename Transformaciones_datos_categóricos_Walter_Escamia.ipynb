```python
import pandas as pd
```

```python
adidas = pd.read_csv("adidas.csv")
```

```python
adidas
```

```python
cols = ['url', 'name', 'sku', 'currency',
       'availability', 'color', 'category', 'source', 'source_website',
       'breadcrumbs', 'description', 'brand', 'images', 'country', 'language']
```

```python
cols
```

```python
cols_validas = []
cols_many_levels = []
cols_one_level = []

size_df = adidas.shape[0]
umbral = .95

for c_ in cols:
    
    niveles = adidas.loc[ : , [c_] ].drop_duplicates().shape[0]
    
    if ( niveles / size_df ) >= umbral:
        cols_many_levels.append(c_)
    elif niveles == 1:
        cols_one_level.append(c_)
    else:
        cols_validas.append(c_)
```

```python
cols_validas
```

```python
size_df = adidas.shape[0]
umbral = .98
cols_buscadas1 = []

for c_ in cols:
    
    niveles = adidas.loc[ : , [c_] ].drop_duplicates().shape[0]
    
    if (niveles > 1) and ( niveles / size_df ) >= umbral:
          cols_buscadas1.append(c_)
```

```python
cols_validas = cols_buscadas1
```

```python
cols_validas
```

```python
size_df = adidas.shape[0]
umbral = .98
cols_buscadas2 = []

for c_ in cols:
    
    niveles = adidas.loc[ : , [c_] ].drop_duplicates().shape[0]
    
    if ( 1 < niveles ) and ( niveles < 6 ):
          cols_buscadas2.append(c_)
```

```python
cols_buscadas2
```

```python
### availability
```

```python
levels_availability = adidas.groupby(["availability"], as_index = False ).agg({"sku":{"count"}})
levels_availability.columns = levels_availability.columns.droplevel(1)
```

```python
levels_availability
```

```python
levels_availability['availability_t'] = 'Others'

levels_availability
```

```python
levels_availability.apply( lambda r: r['availability'] if r['sku'] > 3 else r['availability_t'], axis = 1 )
```

```python
levels_availability['availability_t'] = levels_availability.apply( lambda r: r['availability'] if r['sku'] > 421 else r['availability_t'], axis = 1 )
```

```python
levels_availability
```

```python
levels_availability.loc[ : , ["availability","availability_t"] ]
```

```python
levels_availability = levels_availability.loc[ : , ["availability","availability_t"] ]
```

```python
adidas1 = adidas.merge( levels_availability, on =["availability"], how="left")
```

```python
adidas1
```
